```python
def evaluator(num, expression):
    m = expression

    b = re.sub("[' ';!@#$%^&/\)\].\(:\[,\+\-\*]", "", m)
    try:
        for lin in b:
            int(lin)
    except:
        a = lin

    x = float(num)

    if f'+{a}' in m:
        k = m.replace(f'+{a}', f'+{x}')
    elif f'-{a}' in m:
        k = m.replace(f'-{a}', f'-{x}')
    elif a in m:
        k = m.replace(a, f'*{x}')
    else:
        pass

    if a in k:
        k = k.replace(a, f'*{x}')
    else:
        pass

    if '***' in k:
        k= k.replace('***', '**')
    else:
        pass

    if '(*' in k:
        k = k.replace('(*', '(')
    else:
        pass

    if '*' in k[0]:
        k = k.replace('*', '', 1)
    else:
        pass

    if '+*' in k:
         k = k.replace('+*', '+')
    elif '-*' in k:
         k = k.replace('-*', '-')
    elif '/*' in k:
        k = k.replace('/*', '/')
    else:
        pass

    return eval(k)

print("\nPlease type in the expression")
question = input(f"{Colour.RED}{Colour.BOLD}Please do not input white spaces and do make proper use of brackets{Colour.END}: ")

dd = range(-100, 100)
de = []

for i in dd:
    b = evaluator(i, question)
    de.append(b)
    
plt.plot(dd, de)
```
